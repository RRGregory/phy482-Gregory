```python
import numpy as np                
import matplotlib.pyplot as plt   
%matplotlib inline
```

Plotting EMF vs time for different values of a

```python
w = 1
v = 1
a = 1
x0 = 1
B0 = 1

t = np.arange(0,30,0.1)
EMF = -w*(2*v*a*t + v*B0 + a*x0)
plt.plot(t,EMF,label="a = 1")

a = 10
EMF = -w*(2*v*a*t + v*B0 + a*x0)
plt.plot(t,EMF, label="a = 10")

a = 100
EMF = -w*(2*v*a*t + v*B0 + a*x0)
plt.plot(t,EMF, label = "a = 100")

a = 0.1
EMF = -w*(2*v*a*t + v*B0 + a*x0)
plt.plot(t,EMF, label = "a = 0.1")

plt.legend(loc = 0)
plt.xlabel("time")
plt.ylabel("EMF")
plt.title("Different values of a")
```
