```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def I_amp(w):
    
    V0 = 1
    R = 1
    L = 1
    C = 1
    
    result = V0*np.sqrt((1/R**2) + (1/(w*L) - w*C)**2)
    
    return result
```

```python
w = np.arange(0.1, 7.0, 0.1)
plt.plot(w, I_amp(w))
plt.grid()
plt.xlabel("frequency")
plt.ylabel("Magnitude of Current")
plt.title("Magnitude of current vs frequency")

plt.show()
```

The current is at an extremum when $\omega = \frac{1}{\sqrt{LC}}$
