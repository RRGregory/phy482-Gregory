Positive Point Charge

```python
import numpy as np                # imports the numerical library
import matplotlib.pyplot as plt   # imports the plotting library matplotlib.pyplot.quiver
%matplotlib inline
```

```python
x = np.arange(-5,5,0.5)
y = np.arange(-5,5,0.5)
X, Y = np.meshgrid(x,y)

r = (X**2 + Y**2)**(0.5)

plt.quiver(x,y, X/(r**3), Y/(r**3))
```

Electric Dipole

```python
x = np.arange(-4,4,0.2)
y = np.arange(-4,4,0.2)
X, Y = np.meshgrid(x,y)

plt.streamplot(x, y, (X+1)/(((X+1)**2 + Y**2)**(3/2)) - (X-1)/(((X-1)**2 + Y**2)**(3/2)),
               Y/(((X+1)**2 + Y**2)**(3/2)) - Y/(((X-1)**2 + Y**2)**(3/2)))
```

Electric Quadrupole

```python
plt.streamplot(x,y, (X+1)/(((X+1)**2+(Y+1)**2)**(3/2)) - (X+1)/(((X+1)**2+(Y-1)**2)**(3/2)) + 
               (X-1)/(((X-1)**2+(Y-1)**2)**(3/2)) - (X-1)/(((X-1)**2+(Y+1)**2)**(3/2)) , 
              (Y+1)/(((X+1)**2+(Y+1)**2)**(3/2)) - (Y-1)/(((X+1)**2+(Y-1)**2)**(3/2)) + (Y-1)/(((X-1)**2+(Y-1)**2)**(3/2)) - 
              (Y+1)/(((X-1)**2+(Y+1)**2)**(3/2)))
```
