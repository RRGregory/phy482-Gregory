```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(w):
    R = 1
    L=1
    return R/(np.sqrt((R**2 + (w*L)**2)))

w = np.arange(0.0, 20.0, 0.1)

plt.plot(w,f(w))
plt.xlabel("frequency")
plt.ylabel("|Vout/Vin|")
plt.show()
```

This is a low-pass filter.

```python
t = np.arange(0.0, 20.0, 0.1)

def Vin(t):
    Vin = 0
    k = 1
    w = 1
    
    while k<100:
        Vk = (4/np.pi)*np.sin(w*t*(2*k-1))/(2*k-1)
        Vin = Vin + Vk
        k = k+1

    return Vin

plt.plot(t,Vin(t))
plt.xlabel("t")
plt.ylabel("Vin")
plt.title("Vin vs time")
plt.show()
```

```python
def Vout(t):
    Vout = 0
    k = 1
    w = 1
    L = 1
    R = 1
    
    while k<100:
        Vk = (4/np.pi)*np.sin(w*t*(2*k-1) - 1/np.tan(w*L/R))/(2*k-1)
        Vout = Vout + Vk
        k = k+1

    return Vout

plt.plot(t,Vout(t))
plt.xlabel("t")
plt.ylabel("Vout")
plt.title("Vout vs time")
plt.show()
```
